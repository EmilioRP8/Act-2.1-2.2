```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import erp as erp  # Aseguramos que tu librería 'erp' sea utilizada

# Cargar la base de datos
df = pd.read_csv("listings.csv")

# Contar los valores únicos en la columna 'room_type'
room_type_counts = df['room_type'].value_counts()
print("Tipos de habitación y sus cantidades:")
print(room_type_counts)

# Seleccionar columnas necesarias
extracciondecol = [
    'host_acceptance_rate', 'host_response_rate', 'review_scores_location',
    'review_scores_cleanliness', 'price', 'availability_365', 'number_of_reviews',
    'reviews_per_month', 'review_scores_communication', 'room_type'
]
df = df[extracciondecol]

# Limpiar columnas con porcentajes y convertir a valores numéricos
df['host_acceptance_rate'] = df['host_acceptance_rate'].astype(str).str.replace('%', '', regex=False)
df['host_response_rate'] = df['host_response_rate'].astype(str).str.replace('%', '', regex=False)

# Convertir las columnas de porcentaje a números decimales
df['host_acceptance_rate'] = pd.to_numeric(df['host_acceptance_rate'], errors='coerce') / 100
df['host_response_rate'] = pd.to_numeric(df['host_response_rate'], errors='coerce') / 100

# Limpiar columna de precio
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Identificar y manejar los valores nulos con la librería 'erp'
df = erp.sustitucion_mean(df)  # Usamos la función 'sustitucion_mean' de tu librería para reemplazar los nulos por la media

# Filtrar tipos de habitación
types_selected = ["Private room", "Entire home/apt", "Shared room", "Hotel room"]
df_filtered = df[df['room_type'].isin(types_selected)]

# Gráficos de correlación solicitados y cálculos de R² y correlación
comparisons = [
    ('host_acceptance_rate', 'host_response_rate'),
    ('review_scores_location', 'review_scores_cleanliness'),
    ('host_acceptance_rate', 'price'),
    ('availability_365', 'number_of_reviews'),
    ('host_acceptance_rate', 'number_of_reviews'),
    ('reviews_per_month', 'review_scores_communication')
]

for x, y in comparisons:
    df_pair = df_filtered[[x, y]].dropna().copy()
    model_temp = LinearRegression()
    model_temp.fit(df_pair[[x]], df_pair[y])
    y_pred_temp = model_temp.predict(df_pair[[x]])

    coef_Deter_temp = model_temp.score(df_pair[[x]], df_pair[y])
    coef_Correl_temp = np.sqrt(coef_Deter_temp)
    print(f"{x} vs {y} -> R²: {coef_Deter_temp:.4f}, Correlación: {coef_Correl_temp:.4f}")

    df_pair['Predicciones'] = y_pred_temp
    df_pair = df_pair.sort_values(by=x)

    sns.scatterplot(x=x, y=y, data=df_pair)
    sns.lineplot(x=x, y='Predicciones', data=df_pair, color='red')
    plt.title(f'{x} vs {y} (con línea de predicción)')
    plt.show()

# Seleccionar solo las columnas numéricas para la matriz de correlación
df_numeric = df_filtered.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Generar el Heatmap de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación - Heatmap')
plt.tight_layout()
plt.show()

# Definir variables para regresión lineal múltiple
Vars_Indep = df_filtered[[
    "host_acceptance_rate", "host_response_rate", "price",
    "availability_365", "number_of_reviews", "reviews_per_month"
]]
Var_Dep = df_filtered["review_scores_cleanliness"]

# Eliminar filas con valores nulos
Vars_Indep = Vars_Indep.dropna()
Var_Dep = Var_Dep.loc[Vars_Indep.index]

# Crear y entrenar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X=Vars_Indep, y=Var_Dep)

# Evaluar el modelo general
coef_Deter = model.score(X=Vars_Indep, y=Var_Dep)
print("Regresión Múltiple - Coeficiente de determinación (R²):", coef_Deter)

coef_Correl = np.sqrt(coef_Deter)
print("Regresión Múltiple - Coeficiente de correlación:", coef_Correl)

# Predecir valores
y_pred = model.predict(X=Vars_Indep)

# Insertar predicciones en el DataFrame
df_filtered = df_filtered.loc[Vars_Indep.index].copy()
df_filtered.insert(0, 'Predicciones', y_pred)

# Ordenar por 'host_acceptance_rate' para visualizar bien la línea de predicción
ordenado = df_filtered.sort_values(by='host_acceptance_rate')

# Ajuste de tamaño automático según el número de observaciones
height = max(6, min(0.05 * len(ordenado), 20))
plt.figure(figsize=(10, height))

sns.scatterplot(x='host_acceptance_rate', y='review_scores_cleanliness', color="blue", data=ordenado)
sns.scatterplot(x='host_acceptance_rate', y='Predicciones', color="red", data=ordenado)
sns.lineplot(x='host_acceptance_rate', y='Predicciones', color="red", data=ordenado)
plt.title("Comparación entre valores reales y predichos (regresión múltiple)")
plt.tight_layout()
plt.show()
```
